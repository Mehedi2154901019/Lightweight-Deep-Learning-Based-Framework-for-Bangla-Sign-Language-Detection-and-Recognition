```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import (
    f1_score, precision_score, recall_score,
    classification_report, confusion_matrix
)
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping,ModelCheckpoint
```

```python
# Load dataset
df = pd.read_csv("bdsl47all_merged.csv")
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values  # labels 0-46
```

```python
# One-hot encode labels
encoder = OneHotEncoder(sparse_output=False)
y_onehot = encoder.fit_transform(y.reshape(-1, 1))
```

```python
# Setup cross-validation
k_folds = 5
skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)
```

```python
# Track metrics
fold_accuracies = []
fold_f1_macro = []
fold_f1_weighted = []
fold_precision = []
fold_recall = []
per_class_f1_scores = []
history_per_fold = []
class_labels = [str(i) for i in range(47)]

for fold, (train_idx, val_idx) in enumerate(skf.split(X, y)):
    print(f"\nğŸš€ Fold {fold + 1}/{k_folds}")

    # Split
    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y_onehot[train_idx], y_onehot[val_idx]

    # Scale
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)

    # Build model
    model = Sequential([
        Dense(128, input_shape=(X.shape[1],)),
        BatchNormalization(),
        LeakyReLU(),
        Dense(128),
        BatchNormalization(),
        LeakyReLU(),
        Dense(64),
        BatchNormalization(),
        LeakyReLU(),
        Dropout(0.33),
        Dense(32),
        BatchNormalization(),
        LeakyReLU(),
        Dropout(0.25),
        Dense(47, activation='softmax')
    ])
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    model.summary()
    checkpoint_cb = ModelCheckpoint('Final_MLP.keras', save_best_only=True, monitor='val_accuracy', mode='max', verbose=1)
    early_stop = EarlyStopping(monitor='val_loss', patience=30, restore_best_weights=True)

    history = model.fit(X_train, y_train,
                        validation_data=(X_val, y_val),
                        epochs=500,
                        batch_size=64,
                        callbacks=[checkpoint_cb,early_stop],
                        verbose=1)

    history_per_fold.append(history.history)

    # Evaluate
    score = model.evaluate(X_val, y_val, verbose=0)
    val_accuracy = score[1] * 100
    fold_accuracies.append(val_accuracy)

    y_val_pred = model.predict(X_val)
    y_val_pred_classes = np.argmax(y_val_pred, axis=1)
    y_val_true_classes = np.argmax(y_val, axis=1)

    # Metrics
    f1_macro = f1_score(y_val_true_classes, y_val_pred_classes, average='macro') * 100
    f1_weighted = f1_score(y_val_true_classes, y_val_pred_classes, average='weighted') * 100
    precision = precision_score(y_val_true_classes, y_val_pred_classes, average='macro') * 100
    recall = recall_score(y_val_true_classes, y_val_pred_classes, average='macro') * 100

    fold_f1_macro.append(f1_macro)
    fold_f1_weighted.append(f1_weighted)
    fold_precision.append(precision)
    fold_recall.append(recall)

    # Per-class F1
    f1_per_class = f1_score(y_val_true_classes, y_val_pred_classes, average=None)
    per_class_f1_scores.append(f1_per_class)

    # Print Report
    print(f"âœ… Fold {fold + 1} Accuracy: {val_accuracy:.2f}%")
    print(f"ğŸ“ˆ F1 Macro: {f1_macro:.2f}% | F1 Weighted: {f1_weighted:.2f}%")
    print(f"ğŸ“Œ Precision: {precision:.2f}% | Recall: {recall:.2f}%")
    print("\nğŸ“‹ Classification Report:")
    print(classification_report(y_val_true_classes, y_val_pred_classes, digits=2))

    # Confusion Matrix
    cm = confusion_matrix(y_val_true_classes, y_val_pred_classes)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, cmap='Blues', xticklabels=class_labels, yticklabels=class_labels)
    plt.title(f"Confusion Matrix - Fold {fold + 1}")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()
```

```python
# Summary
print("\nğŸ“Š Cross-validation Summary:")
print(f"âœ… Accuracy Avg: {np.mean(fold_accuracies):.2f}% Â± {np.std(fold_accuracies):.2f}%")
print(f"ğŸ“ˆ F1 Macro Avg: {np.mean(fold_f1_macro):.2f}% Â± {np.std(fold_f1_macro):.2f}%")
print(f"ğŸ“ˆ F1 Weighted Avg: {np.mean(fold_f1_weighted):.2f}% Â± {np.std(fold_f1_weighted):.2f}%")
print(f"ğŸ“Œ Precision Avg: {np.mean(fold_precision):.2f}% Â± {np.std(fold_precision):.2f}%")
print(f"ğŸ“Œ Recall Avg: {np.mean(fold_recall):.2f}% Â± {np.std(fold_recall):.2f}%")
```

```python
# Average per-class F1
mean_per_class_f1 = np.mean(per_class_f1_scores, axis=0)
print("\nğŸ“‹ Average F1 Score per Class (across folds):")
for idx, f1 in enumerate(mean_per_class_f1):
    print(f"Class {idx}: {f1:.4f}")
```

```python
# Accuracy and Loss Plots (separate)
for i, hist in enumerate(history_per_fold):
    # Accuracy Plot
    plt.figure(figsize=(7, 5))
    plt.plot(hist['accuracy'], label='Train Accuracy')
    plt.plot(hist['val_accuracy'], label='Val Accuracy')
    plt.title(f"Accuracy - Fold {i + 1}")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Loss Plot
    plt.figure(figsize=(7, 5))
    plt.plot(hist['loss'], label='Train Loss')
    plt.plot(hist['val_loss'], label='Val Loss')
    plt.title(f"Loss - Fold {i + 1}")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
